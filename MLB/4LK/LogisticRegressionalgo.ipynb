```
import random
import torch
from torch.autograd import Variable
from torch.nn import functional as F
import numpy as np          
from data_process import get_CIFAR10_data
import math
from scipy.spatial import distance
from models import LogisticRegression
from kaggle_submission import output_submission_csv
%matplotlib inline

# You can change these numbers for experimentation
# For submission we will use the default values 
TRAIN_IMAGES = 49000
VAL_IMAGES = 1000
TEST_IMAGES = 5000

data = get_CIFAR10_data(TRAIN_IMAGES, VAL_IMAGES, TEST_IMAGES)
X_train, y_train = data['X_train'], data['y_train']
X_val, y_val = data['X_val'], data['y_val']
X_test, y_test = data['X_test'], data['y_test']
def forward(self,image):
    image = image.reshape(-1,3*32*32)
    out = self.linear(image)
    return out
    
def get_acc(pred, y_test):
    return np.sum(y_test==pred).item()/len(y_test)
```

```
logreg = LogisticRegression.LogisticRegression() 
logreg.train(X_train, y_train)
```

```
X_train = data["X_train"][0]
X_train = np.reshape(X_train,(-1,3*32*32))
X_train = np.array(X_train, dtype=np.float32)
X_train = torch.from_numpy(X_train)

pred_logreg = logreg.predict(X_train)
print('The training accuracy is given by : %f' % pred_logreg, y_train)
```

```
X_val = data["X_val"][0]
X_val = np.reshape(X_val,(-1,3*32*32))
X_val = np.array(X_val, dtype=np.float32)
X_val = torch.from_numpy(X_val)

pred_logreg = logreg.predict(X_val)
print('The validation accuracy is given by : %f' % pred_logreg, y_val)
```

```
X_test = data["X_test"][0]
X_test = np.reshape(X_test,(-1,3*32*32))
X_test = np.array(X_test, dtype=np.float32)
X_test = torch.from_numpy(X_test)

pred_logreg = logreg.predict(X_test)
print('The testing accuracy is given by : %f' % pred_logreg, y_test)
```
